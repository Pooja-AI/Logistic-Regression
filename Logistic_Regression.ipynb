### Logit Function<br>
<font size = "3">
Given a probability p, the corresponding odds are calculated as p / (1 – p). For example if p=0.75, the odds are 3 to 1: 0.75/0.25 = 3<br><br><br>
The logit function is simply the logarithm of the odds: logit(x) = log(x / (1 – x))<br>
The value of the logit function heads towards infinity as p approaches 1 and towards negative infinity as it approaches 0<br><br>
The logit function is useful in analytics because it maps probabilities (which are values in the range [0, 1]) to the full range of real numbers. In particular, if you are working with “yes-no” (binary) inputs it can be useful to transform them into real-valued quantities prior to modeling. This is essentially what happens in logistic regression.<br><br>
The formula for the sigmoid function is σ(x) = 1/(1 + exp(-x))<br>

### Notes In Logistic Regression

<li>A binary dependent variable can have only two values, like 0 or 1, win or lose, pass or fail, healthy or sick, etc</li>

### Optimization

<font size="3">
A logistic regression model or, more specifically, its weight vector w is learned from the training data, <br><br>

with the goal of predicting a positive sample as close to 1 as possible and predicting a negative sample as close to 0 as possible. <br><br>

In mathematical language, the weights are trained so as to minimize the cost defined as the mean squared error (MSE), which measures the average of squares of difference between the<b color="blue"> truth and the prediction.</b><br><br>

### Convex and Non Convex

- A convex optimization problem maintains the properties of a linear programming problem and a non convex problem the properties of a non linear programming problem

- The basic difference between the two categories is that in 
- convex optimization there can be only one optimal solution, which is <b>globally optimal</b> or you might prove that there is no feasible solution to the problem, while in 
- nonconvex optimization may have multiple<b> locally optimal points </b>and it can take a lot of time to identify whether the problem has no solution or if the solution is global.
- Hence, the efficiency in time of the convex optimization problem is much better. From my experience a convex problem usually is much more easier to deal with in comparison to a non convex problem which takes a lot of time and it might lead you to a dead end.

### What is Logistic Regression<br>

<font size="3" color="blue">Logistic regression is the appropriate regression analysis to conduct when the dependent variable is dichotomous (binary).  Like all regression analyses, the logistic regression is a predictive analysis.  Logistic regression is used to describe data and to explain the relationship between one dependent binary variable and one or more nominal, ordinal, interval or ratio-level independent variables.</font>

### cost function / Loss Function

- If y(i)=1, when it predicts correctly (positive class in 100% probability), the sample cost j is 0; the cost keeps increasing when it is less likely to be the positive class; when it incorrectly predicts that there is no chance to be the positive class, the cost is infinitely high.
<br><br>
- If y(i)=1, when it predicts correctly (positive class in 100% probability), the sample cost j is 0; the cost keeps increasing when it is less likely to be the positive class; when it incorrectly predicts that there is no chance to be the positive class, the cost is infinitely high.
<br>

Minimizing this alternative cost function is actually equivalent to minimizing the MSE-based cost function. The advantages of choosing it over the other one include the following:
